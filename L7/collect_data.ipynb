```python
import requests
from bs4 import BeautifulSoup
import os
import random

def download_images(keyword, n1, n2):
    count = 1
    while count <= n1:
        page = random.randint(0, 350)
        url = f"https://www.bing.com/images/search?q={keyword}&first={page}"
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        image_links = soup.find_all('img', class_='mimg')

        i = 0
        while count <= n1 and i < len(image_links):
            image_link = image_links[i]
            i += 1
            
            try:
                image_url = image_link['src']
                image_content = requests.get(image_url).content
                with open(f"data/train/{keyword}/{keyword}_{count}.jpg", 'wb') as f:
                    f.write(image_content)
                if count % 10 == 9:
                    print(f"{count + 1} / {n1}")
                count += 1
            except:
                pass
        
    count = 1
    while count <= n2:
        page = random.randint(0, 350)
        url = f"https://www.bing.com/images/search?q={keyword}&first={page}"
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        image_links = soup.find_all('img', class_='mimg')

        i = 0
        while count <= n2 and i < len(image_links):
            image_link = image_links[i]
            i += 1
            
            try:
                image_url = image_link['src']
                image_content = requests.get(image_url).content
                with open(f"data/val/{keyword}/{keyword}_{count}.jpg", 'wb') as f:
                    f.write(image_content)
                if count % 10 == 9:
                    print(f"{count + 1} / {n2}")
                count += 1
            except:
                pass

if __name__ == "__main__":
    keywords = ["Macaca", "Tarsiidae", "Rhinopithecus", "Mandrillus sphinx", "Callitrichinae", "Cebus capucinus", "Nasalis larvatus", "Hylobatidae"]
    
    for k in keywords:
        if not os.path.exists(f"data/train/{k}"):
            os.makedirs(f"data/train/{k}")
        if not os.path.exists(f"data/val/{k}"):
            os.makedirs(f"data/val/{k}")
            
        print(k)
        download_images(k, 200, 20)
```
